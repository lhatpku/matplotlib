## Adjusting Subplots

In the code below we use the `subplots` function to create a 3x2 grid. 

```python
plt.subplots(3, 2);
```

### Tight Layout

```python
plt.subplots(3, 2)
plt.tight_layout();
```

### `pyplot.subplots_adjust`

```python
plt.subplots(3, 2)
plt.subplots_adjust(hspace=0.5);
```
